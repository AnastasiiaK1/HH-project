Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)

* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def job_expirience(exp):
  if exp =='Не указано' or exp is np.nan:
    return np.nan
  year_list = ['лет','год','года']
  month_list = ['месяц','месяца','месяцев']
  total_exp = 0
  exp_list = exp.split(' ')[2:6]
  if exp_list[1] in year_list and exp_list[3] in month_list:
    total_exp = int(exp_list[0])*12 + int(exp_list[2])
  elif exp_list[1] in year_list and exp_list[3] not in month_list:
    total_exp = int(exp_list[0])*12
  elif exp_list[1] in month_list:
    total_exp = int(exp_list[0])
  return total_exp
```

```python
data['Опыт работы (месяц)']= data['Опыт работы'].apply(job_expirience)
data = data.drop('Опыт работы',axis = 1)
```

Функция вычленения города

```python
def city(info):
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск',
                      'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    result_spl = info.split(' ')[0]
    if result_spl == 'Москва':
      return 'Москва'
    elif result_spl == 'Санкт-Петербург':
      return 'Санкт-Петербург'
    elif result_spl in million_cities:
      return 'город-миллионник'
    else:
      return 'другие'
```

функция  "Готовность к переезду"

```python
def ready_to_move(info):
  pas_pret = ['не готова к переезду', 'не готов к переезду']
  if pas_pret[0] in info or pas_pret[1] in info:
    return False
  else:
    return True
```

Функция "Готовность к командировкам"

```python
def ready_for_trip(info):
  ne_gotov = ['не готова к командировкам', 'не готов к командировкам']
  if ne_gotov[0] in info or ne_gotov[1] in info or 'командировка' not in info:
    return False
  else:
    return True
```

```python
data['Город']=data['Город, переезд, командировки'].apply(city)
data['Город']=data['Город'].astype('category')
data['Готовность к переезду']= data['Город, переезд, командировки'].apply(ready_to_move)
data['Готовность к командировкам']=data['Город, переезд, командировки'].apply(ready_for_trip)
```

```python
data = data.drop('Город, переезд, командировки',axis = 1)
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
log_age = np.log(column3['Возраст'])

plt.figure(figsize=(10, 6))
histplot = sns.histplot(log_age, bins=30, kde=True, color='blue', alpha=0.7)
histplot.axvline(log_age.mean(), color='k', lw=2, label='Среднее')
histplot.axvline(log_age.mean()-3*log_age.std(), color='g', lw=2, linestyle='--', label='-3 сигмы')
histplot.axvline(log_age.mean()+4*log_age.std(), color='r', lw=2, linestyle='--', label='+4 сигмы')

plt.title('Распределение признака "возраст" в логарифмическом масштабе')
plt.xlabel('Логарифм возраста')
plt.ylabel('Частота')
plt.legend()
plt.grid(True)
plt.show()
```

```python
mean_log = log_age.mean()
std_log = log_age.std()
z_score = (log_age - mean_log)/std_log
mask_out = (z_score < -3)|(z_score > 4)
out_df = column3[mask_out]
clean_df = column3[~mask_out]
print('выбросы')
print(out_df['Логарифм возраста '])
```

Нашла выбросы с помощью метода z-отклонения и удалила их из данных - записав в датафрейм clean_df, используя логарифмический масштаб, сделала послабление на 1 сигму (взяла 4 сигмы) в правую сторону.
